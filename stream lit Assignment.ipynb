```python
%pip install streamlit
import streamlit as st

st.set_page_config(page_title="Calculator App", page_icon="🤓☝️")

st.title("🧮 Calculator App")
st.write("Simple Calculator deployed on Streamlit")

# Inputs
num1 = st.number_input("Enter first number", value=0.0)
num2 = st.number_input("Enter second number", value=0.0)

operation = st.radio(
    "Select Operation",
    ("Addition", "Subtraction", "Multiplication", "Division")
)

if st.button("Calculate"):
    if operation == "Addition":
        st.success(f"Result: {num1 + num2}")

    elif operation == "Subtraction":
        st.success(f"Result: {num1 - num2}")

    elif operation == "Multiplication":
        st.success(f"Result: {num1 * num2}")

    elif operation == "Division":
        if num2 == 0:
            st.error("❌ Division by zero not allowed")
        else:
            st.success(f"Result: {num1 / num2}")
```

```python
import streamlit as st

st.title("Student Information Form")

# Text input for name
name = st.text_input("Enter your name")

# Number input for age
age = st.number_input("Enter your age", min_value=0, max_value=100, step=1)

# Dropdown selection
course = st.selectbox(
    "Select Course",
    ["BCA", "BBA", "MBA", "MCA", "Other"]
)

# Conditional input: If "Other" is selected, show an extra text input
if course == "Other":
    custom_course = st.text_input("Please specify your course")
    # Update the course variable with the custom input if provided
    if custom_course:
        course = custom_course

# Radio button
gender = st.radio(
    "Select Gender",
    ["Male", "Female", "Other"]
)

# Multiselect for skills
skills = st.multiselect(
    "Select your skills",
    ["Python", "Java", "C++", "AI", "Machine Learning"]
)

# Slider input
marks = st.slider(
    "Enter Marks",
    0, 100
)

# Checkbox
agree = st.checkbox(
    "I agree to terms"
)

# Button
if st.button("Submit"):
    if agree:
        st.write("### Submitted Details:")
        st.write("**Name:**", name)
        st.write("**Age:**", age)
        st.write("**Course:**", course)
        st.write("**Gender:**", gender)
        st.write("**Marks:**", marks)
        st.write("**Skills:**", ", ".join(skills) if skills else "None")
        
        st.success("Form Submitted Successfully!")
    else:
        st.error("Please accept the terms and conditions before submitting.")
```

```python
# Import Streamlit library and give it a short name "st"
import streamlit as st


# Creates the main title on webpage
st.title("BMI Calculator")


# Creates a number input box for user weight
# Value entered is stored in variable "weight"
weight = st.number_input("Weight (kg)")


# Creates another number input for height
# Value entered is stored in variable "height"
height = st.number_input("Height (m)")


# Creates a button named "Calculate BMI"
# Code inside this block runs only after button click
if st.button("Calculate BMI"):


    # BMI Formula:
    # BMI = Weight / Height²
    # **2 means square of height
    bmi = weight / (height**2)


    # Display BMI result in a green success message
    # :.2f shows only 2 digits after decimal
    st.success(f"BMI: {bmi:.2f}")


    # If BMI is below 18.5
    # Show Underweight warning
    if bmi < 18.5:
        st.warning("Underweight")


    # If BMI is less than 25
    # Show Normal category
    elif bmi < 25:
        st.success("Normal")


    # If BMI is above or equal to 25
    # Show Overweight category
    else:
        st.error("Overweight")
```

```python
# Import Streamlit library
import streamlit as st


# Create app title
st.title("Currency Converter")


# Input amount from user
amount = st.number_input(
    "Enter Amount",
    min_value=0.0
)


# Dropdown to select source currency
from_currency = st.selectbox(
    "From",
    ["INR", "USD", "EUR"]
)


# Dropdown to select target currency
to_currency = st.selectbox(
    "To",
    ["INR", "USD", "EUR"]
)


# Sample conversion rates
rates = {
    ("INR","USD"):0.012,
    ("USD","INR"):83,
    ("INR","EUR"):0.011,
    ("EUR","INR"):90,
    ("USD","EUR"):0.92,
    ("EUR","USD"):1.08
}


# Convert button
if st.button("Convert"):


    # If both currencies same
    if from_currency == to_currency:
        result = amount


    # Otherwise use conversion rate
    else:
        result = amount * rates[(from_currency,to_currency)]


    # Show result
    st.success(
        f"{amount} {from_currency} = {result:.2f} {to_currency}"
    )
```

```python
# Import Streamlit library and give it short name "st"
import streamlit as st


# Creates main heading on webpage
st.title("To Do App")


# Check whether "tasks" list already exists
# in Streamlit session memory
if "tasks" not in st.session_state:

    # Create an empty list for storing tasks
    st.session_state.tasks=[]


# Create a text box to type task
# Store entered text inside variable task
task=st.text_input("Task")


# Create Add button
# Code runs only when button is clicked
if st.button("Add"):

    # Add entered task into task list
    st.session_state.tasks.append(task)


# Loop through every task in task list
for t in st.session_state.tasks:

    # Display task with check mark
    st.write("✅",t)
```

```python
import streamlit as st

st.title("Password Checker")

pwd=st.text_input(
    "Password",
    type="password"
)

if st.button("Check"):

    score=0

    if len(pwd)>=8:
        score+=1

    if any(c.isdigit() for c in pwd):
        score+=1

    if any(c.isupper() for c in pwd):
        score+=1

    st.write("Strength:",score,"/3")
```
