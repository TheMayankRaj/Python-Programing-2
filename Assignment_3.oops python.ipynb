# Write a program to iterate over a list using an iterator and a while loop. [10, 20, 30, 40]

```python
number=[10,20,30,40]
num_iterator=iter(number)
while True:
    try:
        print(next(num_iterator))
    except StopIteration:
        break
```

# write a python program to grnerate fibonacci series using a generator function and print the first ten numbers 


```python
def fibonacci(n):
    a, b = 0, 1
    for n in range(n):
        yield a
        a, b = b, a + b

fib = fibonacci(5)
print(list(fib))  
```

# write a python program to create a generator that performs a countdown and demonstrate the use of next() and send() to control the generator.

```python
def countdown(n):
    while n > 0:
        val = (yield n)
        if val is not None:
            n = val
        else:
            n -= 1

cd = countdown(5)
print(next(cd))   
print(cd.send(3)) 
print(next(cd))
```

# write a python program to demonstrate the use of the decoratro that exicute the code before and after a function

```python
def my_decorator(func):
    def wrapper():
        print("Something is happening before the function is called.")
        func()
        print("Something is happening after the function is called.")
    return wrapper

@my_decorator
def say_hello():
    print("Hello!")

say_hello()
```

# write a python program to create a decorator that logs a functions name and its arguunments beefore executing the function

```python
def log_decorator(func):
    def wrapper(a, b):
        print(f"Calling {func.__name__} with arguments: {a}, {b}")
        return func(a, b)
    return wrapper

@log_decorator
def multiply(x, y):
    return x * y

print(multiply(4, 5))
```

# write a python progmram to create a decorator that measures and prints the executions time of the functions 

```python

import time

def timer_decorator(func):
    def wrapper():
        start = time.time()
        result = func()
        end = time.time()
        print(f"Time taken: {end - start:.4f} seconds")
        return result
    return wrapper

@timer_decorator
def do_work():
    time.sleep(1)
    print("Work done.")

do_work()
```

# Create a memoization decorator for a function that takes one integer parameter.

```python
def decor(func):
    cache = {}
    def wrapper(n):
        if n in cache:
            return cache[n]
        result = func(n)
        cache[n] = result
        return result
    return wrapper

@decor
def factorial(n):
    if n == 0:
        return 1
    return n * factorial(n - 1)

print(factorial(5))
```
