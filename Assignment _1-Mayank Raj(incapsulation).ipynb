```python
class RacingCar:
    
    def __init__(self):
        self.__speed = 0  
        
    def set_speed(self, speed):
        if speed >= 0:
            self.__speed = speed
        print("Speed is set to:", self.__speed, "m/s")

    def acceleration(self, total_time):
        if total_time > 0:
            acc = self.__speed / total_time
            print("Acceleration is:", acc, "m/s²")
        else:
            print("Time must be greater than 0")

    def braking(self, brake_value):
        if brake_value > 0:
            self.__speed -= brake_value
            if self.__speed < 0:
                self.__speed = 0
        print("Speed after braking:", self.__speed, "m/s")


car = RacingCar()
car.set_speed(120)         
car.acceleration(12)       
car.braking(50)
```

# question_2

```python
class Diet:

    def __init__(self, calories=0):
        self.__calories = calories   

    def junk_calories(self, junk):
        if junk > 0:
            self.__calories = self.__calories + junk
        print("Current Calories:", self.__calories)

    def exercise(self, burn):
        if burn > 0:
            if burn >= self.__calories:
                self.__calories = 0
            else:
                self.__calories = self.__calories - burn
        print("Current Calories:", self.__calories)

    def check_health(self):
        if self.__calories > 30:
            print("Unhealthy")
        else:
            print("Healthy")


d = Diet()
d.junk_calories(40)
d.check_health()

d.exercise(20)
d.check_health()
```

```python
class BankAccount:
    def __init__(self, balance):
        self.__balance = balance

    def get_balance(self):
        return self.__balance

    def deposit(self, amount):
     if amount > 0:
        self.__balance += amount
        print("Amount deposited")
     else:
        print("Invalid deposit")

    def withdraw(self, amount):
     if amount <= self.__balance:
        self.__balance -= amount
        print("Withdrawal successful")
     else:
        print("Insufficient balance")

acc = BankAccount(1000)
acc.deposit(500)       
acc.withdraw(200)       
print(acc.get_balance())
```

```python
class IMBD:
    
    def __init__(self, rating):
        self.__rating = rating  
        print("Initial rating:", self.__rating)

    def new_rating(self, new_rating):
        if new_rating > self.__rating:
            self.__rating = new_rating
            print("Rating increased to:", self.__rating)
        elif new_rating < self.__rating:
            self.__rating = new_rating
            print("Rating decreased to:", self.__rating)
        else:
            print("Rating remains the same:", self.__rating)

    def show_rating(self):
        print("Current rating:", self.__rating)


movie = IMBD(7.0)     
movie.new_rating(8.5) 
movie.new_rating(6.0)   
movie.show_rating()
```

```python
class Book:
    def __init__(self, title, author):
        self.__title = title
        self.__author = author

    def get_title(self):
        return self.__title

    def get_author(self):
        return self.__author

book1 = Book("The Alchemist", "Paulo Coelho") 

print("Title:", book1.get_title())
print("Author:", book1.get_author())
```

```python
class Laptop:
    def __init__(self, price):
        self.__price = price
        
    def set_price(self, new_price):
        if new_price > 0:
            self.__price = new_price
        else:
            print("Price must be positive.")
            
    def get_price(self):
        return self.__price

laptop1 = Laptop(800)  
print("Original Price:", laptop1.get_price())
laptop1.set_price(950)  
print("Updated Price:", laptop1.get_price())
```
