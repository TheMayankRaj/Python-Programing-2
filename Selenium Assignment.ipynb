```python
# COMPLETE SELENIUM CODE

# Step 1: Install Chrome
!wget -q https://dl.google.com/linux/direct/google-chrome-stable_current_amd64.deb
!apt -qq install -y ./google-chrome-stable_current_amd64.deb

# Step 2: Install required libraries
!pip install -q selenium webdriver-manager

# Step 3: Import libraries
#for driving the browser
from selenium import webdriver #Main module to control browser
from selenium.webdriver.chrome.service import Service #Handles ChromeDriver service
from webdriver_manager.chrome import ChromeDriverManager #Automatically installs correct ChromeDriver

#for segregation and typing
from selenium.webdriver.common.by import By #Used to locate elements (id, name, css, etc.)
from selenium.webdriver.common.keys import Keys #Used for keyboard actions (ENTER, ESC, etc.)

#for operations and time applications
from selenium.webdriver.chrome.options import Options #Used to configure browser settings
import time #Used to pause execution (sleep)
import pandas as pd #Used to store and save data (CSV)


# Step 4: Set Chrome options
options = Options()
options.add_argument("--headless=new")   # run in background
options.add_argument("--no-sandbox")     #dont open browser in real time
options.add_argument("--disable-dev-shm-usage")

# Step 5: Setup driver
driver = webdriver.Chrome(
    service=Service(ChromeDriverManager().install()),
    options=options
)

# Step 6: Open Bing
driver.get("https://www.bing.com/")
time.sleep(3)

# Step 7: Search
search_box = driver.find_element(By.NAME, "q")
search_box.send_keys("Final")
search_box.send_keys(Keys.RETURN)

time.sleep(5)
print(driver.page_source)

# Step 8: Print page title
results = driver.find_elements(By.CSS_SELECTOR, "jhk.b_algo")

data = []

for r in results[:5]:   # top 5 results
    try:
        title = r.find_element(By.TAG_NAME, "h2").text
        link = r.find_element(By.TAG_NAME, "a").get_attribute("href")
        data.append([title, link])
    except:
        continue

# Step 9: Print results
print("Top 5 Results:\n")
for i, d in enumerate(data, 1):
    print(f"{i}. {d[0]}")
    print(d[1])
    print()

# Step 10: Save to CSV
df = pd.DataFrame(data, columns=["Title", "Link"])
df.to_csv("bing_results.csv", index=False)

print("✅ Saved as bing_results.csv")

# Step 11: Take screenshot
driver.save_screenshot("bing_page.png")
print("📸 Screenshot saved as bing_page.png")

# Step 12: Close browser
driver.quit()
```

```python
# 🔹 Import required libraries
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager

# 🔹 Start browser
driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()))  # Launch Chrome with auto driver

# 🔹 Open website
driver.get("https://www.bing.com/")  # Open a URL in browser

# 🔹 Navigation
driver.back()        # Go to previous page
driver.forward()     # Go to next page
driver.refresh()     # Reload current page

# 🔹 Close browser
driver.close()       # Close current tab
driver.quit()        # Close entire browser session

# 🔹 Locate elements
from selenium.webdriver.common.by import By

driver.find_element(By.ID, "id")                     # Find element by ID
driver.find_element(By.NAME, "name")                 # Find by name attribute
driver.find_element(By.CLASS_NAME, "class")          # Find by class name
driver.find_element(By.TAG_NAME, "tag")              # Find by tag name
driver.find_element(By.LINK_TEXT, "text")            # Find link with exact text
driver.find_element(By.PARTIAL_LINK_TEXT, "text")    # Find link with partial text
driver.find_element(By.CSS_SELECTOR, "css")          # Find using CSS selector
driver.find_element(By.XPATH, "xpath")               # Find using XPath
driver.find_elements(By.CLASS_NAME, "item")          # Find multiple elements (list)

# 🔹 Element actions
element = driver.find_element(By.ID, "id")  # Store element

element.click()           # Click on element
element.send_keys("Hi")   # Type text into input box
element.clear()           # Clear text from input
element.submit()          # Submit form

# 🔹 Keyboard actions
from selenium.webdriver.common.keys import Keys

element.send_keys(Keys.RETURN)     # Press Enter key
element.send_keys(Keys.CONTROL, 'a')  # Select all text

# 🔹 Waits
driver.implicitly_wait(10)  # Wait up to 10 sec for elements (global wait)

from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC

WebDriverWait(driver, 10).until(
    EC.presence_of_element_located((By.ID, "id"))  # Wait until element appears
)

# 🔹 Get data
element.text                         # Get visible text of element
element.get_attribute("href")       # Get attribute value
driver.title                        # Get page title
driver.current_url                  # Get current URL
driver.page_source                  # Get full HTML source

# 🔹 Window handling
driver.current_window_handle        # Get current window ID
driver.window_handles               # Get all open windows

driver.switch_to.window(driver.window_handles[0])  # Switch to window
driver.switch_to.new_window('tab')  # Open new tab

# 🔹 Scrolling
driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")  # Scroll to bottom

# 🔹 Run JavaScript
driver.execute_script("alert('Hello')")  # Run JS code in browser

# 🔹 Frames / iFrames
driver.switch_to.frame("frame_name")   # Switch to frame
driver.switch_to.default_content()     # Back to main page

# 🔹 Cookies
driver.get_cookies()                      # Get all cookies
driver.add_cookie({"name": "test", "value": "123"})  # Add cookie
driver.delete_all_cookies()               # Delete all cookies

# 🔹 Screenshot
driver.save_screenshot("page.png")  # Take screenshot

# 🔹 Mouse actions
from selenium.webdriver.common.action_chains import ActionChains

actions = ActionChains(driver)
actions.move_to_element(element).click().perform()  # Move & click

# 🔹 Dropdown
from selenium.webdriver.support.ui import Select

select = Select(element)                     # Create dropdown object
select.select_by_visible_text("Option")      # Select by text
select.select_by_index(1)                    # Select by index
select.select_by_value("value")              # Select by value

# 🔹 Alerts
alert = driver.switch_to.alert  # Switch to alert popup
alert.accept()                  # Click OK
alert.dismiss()                 # Click Cancel
alert.text                      # Get alert text

# 🔹 Timeout
driver.set_page_load_timeout(10)  # Max time to load page

# 🔹 Element state
element.is_displayed()   # Check if visible
element.is_enabled()     # Check if enabled
element.is_selected()    # Check if selected (checkbox/radio)

# 🔹 Force click using JS
driver.execute_script("arguments[0].click();", element)  # Click via JS
```
