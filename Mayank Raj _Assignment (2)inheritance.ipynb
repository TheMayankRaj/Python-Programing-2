```python
class Laptop:
    def __init__(self, brand, processor):
        self.brand = brand
        self.processor = processor

class Battery(Laptop):
    def __init__(self, brand, processor, battery_life):
        super().__init__(brand, processor)
        self.battery_life = battery_life

s1 = Battery("Dell", "i9", "12 hours") 
print(s1.brand, s1.processor, s1.battery_life)
```

```python
class Car:
    def __init__(self, model, brand):
        self.model = model
        self.brand = brand
class ElectricCar(Car):
    def __init__(self, model, brand, battery_size):
        super().__init__(model, brand)
        self.battery_size = battery_size
car1 = ElectricCar("Model S", "Tesla", "100KW")
print(car1.model, car1.brand, car1.battery_size)
```

```python

class Computer:
    def __init__(self, brand, processor):
        self.brand = brand
        self.processor = processor
class Laptop(Computer):
    def __init__(self, brand, processor, battery_life):
        super().__init__(brand, processor)
        self.battery_life = battery_life
s1 = Laptop("Samsung Galaxy S24", "512GB", "Android 14") 
print(s1.brand, s1.processor, s1.battery_life)

```

```python
class Artist:
    def __init__(self, name, age):
        self.name = name
        self.age = age
class Musician(Artist):
    def __init__(self, name, age, instrument_played):
        super().__init__(name, age)
        self.instrument_played = instrument_played
x = Musician("Ariana Grande", 31, "Vocals")
print(x.name, x.age, x.instrument_played)
x2 = Musician("Bruno Mars", 38, "Piano")
print(x2.name, x2.age, x2.instrument_played)


```

```python
class Book:
    def __init__(self, writer, name):
        self.writer = writer
        self.name = name
class FictionBook(Book):
    def __init__(self, writer, name, genre):
        super().__init__(writer, name)
        self.genre = genre
s1 = FictionBook("George Orwell", "1984", "Dystopian") 
print(s1.writer, s1.name, s1.genre)
```

```python
class Student:
    def __init__(self, name, marks1, marks2, marks3):
        self.name = name
        self.marks = [marks1, marks2, marks3]

a1 = Student("Ananya", 85, 90, 78)  

total = sum(a1.marks)
length = len(a1.marks)
average = total / length

print(f"Total marks => {total} and its average is {average}")
```
